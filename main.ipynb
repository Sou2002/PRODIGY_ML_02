### **Importing necessary libraries**

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

### **Importing the dataset**

```python
df = pd.read_csv('datasets/Mall_customers.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### **Feature Extraction (Based on Task Criteria)**

```python
df.drop(['CustomerID', 'Gender', 'Age'], axis = 'columns', inplace = True)
```

```python
df.head()
```

### **Outlier Detection**

```python
df.skew()
```

```python
df['Annual Income (k$)'].plot(kind = 'box')
```

```python
df['Spending Score (1-100)'].plot(kind = 'box')
```

### **Scaling**

```python
scaler = StandardScaler()
```

```python
scaled_array = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_array, columns = df.columns)
```

```python
scaled_df.head()
```

### **Finding elbow point**

```python
sse = []
k_rng = range(1, 11)
for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit(df)
    sse.append(km.inertia_)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(range(1, 11), sse)
plt.plot(range(1, 11), sse, linewidth = 2, color = "red", marker = "o")
plt.xlabel("No. of Clusters")
plt.xticks(np.arange(1, 11, 1))
plt.ylabel("SSE")
plt.show()
```

### **Model Training**

```python
km = KMeans(n_clusters = 5)
y = km.fit_predict(scaled_df)
```

```python
scaled_df["label"] = y
```

```python
scaled_df.head()
```

### **Results**

```python
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', hue = "label", palette = ['green', 'orange', 'brown', 'dodgerblue', 'red'], legend = 'full', data = scaled_df, s = 60)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)') 
plt.title('Spending Score (1-100) vs Annual Income (k$)')
plt.show()
```
